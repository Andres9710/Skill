```python
import matplotlib.pyplot as plt
import numpy as np

years = np.array([1993, 1994, 1995])
hotel_A = np.array([450, 650, 650])
hotel_B = np.array([225, 425, 125])
hotel_C = np.array([245, 345, 445])

bar_width = 0.25
bar_positions_A = np.arange(len(years))
bar_positions_B = bar_positions_A + bar_width
bar_positions_C = bar_positions_B + bar_width

plt.figure(figsize=(10, 6))

plt.bar(bar_positions_A, hotel_A, width=bar_width, label='Hotel A')
plt.bar(bar_positions_B, hotel_B, width=bar_width, label='Hotel B')
plt.bar(bar_positions_C, hotel_C, width=bar_width, label='Hotel C')

plt.xlabel('Ao')
plt.ylabel('Valor')
plt.title('Comparacin de Hoteles a lo largo de los Aos (Grfico de Barras)')
plt.xticks(bar_positions_B, years)
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

years = [1993, 1994, 1995]
hotel_A = [450, 475, 390]
hotel_B = [225, 240, 205]
hotel_C = [245, 255, 345]

plt.figure(figsize=(10, 6))

plt.plot(years, hotel_A, marker='o', label='Hawaiian Club', color='blue', linestyle='-', linewidth=2)
plt.plot(years, hotel_B, marker='o', label='French Riviera', color='green', linestyle='-', linewidth=2)
plt.plot(years, hotel_C, marker='o', label='Bahamas Beach', color='orange', linestyle='-', linewidth=2)

plt.xlabel('Ao')
plt.title('Evolucin de Ingresos por Hotel', ha='right')
plt.grid(False)

# Ajustar distancia horizontal y vertical (cambiar los valores segn sea necesario)
horizontal_offset = 0.03
vertical_offset = 10

# Agregar etiquetas directamente sobre los puntos de inters
for x, y_a, y_b, y_c in zip(years, hotel_A, hotel_B, hotel_C):
    plt.text(x + 0.02, y_a + 3, f'{y_a}', ha='right', va='bottom', fontsize=8, color='blue')
    plt.text(x + 0.02, y_b - 4, f'{y_b}', ha='right', va='top', fontsize=8, color='green')
    plt.text(x, y_c + 2, f'{y_c}', ha='right', va='bottom', fontsize=8, color='orange')

# Agregar etiquetas al final de cada lnea
plt.text(years[-1] + horizontal_offset, hotel_A[-1], 'Hawaiian Club', ha='left', va='bottom', fontsize=8, color='blue')
plt.text(years[-1] + horizontal_offset, hotel_B[-1], 'French Riviera', ha='left', va='top', fontsize=8, color='green')
plt.text(years[-1] + horizontal_offset, hotel_C[-1], 'Bahamas Beach', ha='left', va='bottom', fontsize=8, color='orange')

# Mover la etiqueta del eje y a la esquina superior izquierda
plt.text(min(years) - 0.3, max(max(hotel_A), max(hotel_B), max(hotel_C)) + 30, "Ingresos \n'000$", ha='left', va='top', rotation='horizontal', fontsize=10)

# Quitar el borde de la derecha de la grfica
plt.gca().spines['right'].set_visible(False)

# Ajustar el lmite superior del eje Y
plt.ylim(top=500)
plt.xticks(years)

# Guardar la figura en un archivo PDF
plt.savefig("grafico_ingresos_hoteles.pdf")

plt.show()
```
